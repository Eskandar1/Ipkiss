<h1>Implantação do Haar Cascade na Webcam</h1>

<p>Antes de executar as importações verifique em há esses módulos em seu computador, já que algumas precisam de instalação
</p>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from io import BytesIO
from IPython.display import Image, display, clear_output
from PIL import Image as Img
```

<p>Também se o .xml do HaarCascade específico para faces está instalado, nesse caso ele está na pasta classificadores</p>

```python
classificador = cv2.CascadeClassifier('classificadores/haarcascade_frontalface_default.xml')
```

<p>Foi criado três funções para que seja possível a exibição da detecção de faces em tempo real <br/>
    <b>Padronizar imagem</b> recebe um frame do vídeo, e modifica a cor dele, para precaver que uma imagem pode vir com o formato BGR(Blue Green Red) e transforma para o formato padrão RGB(Red Green Blue) e redimensiona o frame para 500x400 pixels, assim todas as imagens vão ser do mesmo tamanho.
</p>

```python
def padronizar_imagem(frame):
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame = cv2.resize(frame , (500,400))
    return frame
```

```python
def detectar(frame):
    imagem_gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
    faces = classificador.detectMultiScale(imagem_gray, 1.3, 5)
    for (x, y, largura, altura) in faces:
        cv2.rectangle(frame,(x,y), (x+largura, y+altura), (255,0,0), 2)
    return frame
```

```python
def exibir_video(frame):
    img = Img.fromarray(frame, "RGB")
    buffer = BytesIO()
    img.save(buffer,format="JPEG")
    display(Image(data=buffer.getvalue()))
    clear_output(wait=True)
```

```python
video = cv2.VideoCapture(0)
try:
    while(True):
        captura_ok, frame = video.read()
        if captura_ok:
            frame = padronizar_imagem(frame)
            frame = detectar(frame)
            exibir_video(frame)
except KeyboardInterrupt:
    video.release()
    cv2.destroyAllWindows()
```
